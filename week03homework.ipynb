```python
#Using np.argmax() and np.unique() function,mode is calculated. np.unique() stored the unique values from horsepower 
#inside mode array and number of times each value is repeated inside the counts array. np.argmax() calculated max value 
#inside the counts array and stores in index variable. Finally printing the vaiue at index of the mode array. 

mode,counts = np.unique(horsepower, return_counts=True)
index = np.argmax(counts)

print(horsepower)
print("Sum of Horsepower: ", np.sum(horsepower))
print("Minium of Horsepower: ", np.min(horsepower))
print("Maximum of Horsepower: ", np.max(horsepower))
print("Median of Horsepower: ", np.median(horsepower))
#optional
print("Mode of Horsepower: ", mode[index]) 
```
